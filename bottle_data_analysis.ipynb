```python
bottle_dataset.isna().sum()
```

```python
bottle_dataset.drop('Unnamed: 0', axis = 1, inplace=True)
bottle_dataset_columns = bottle_dataset.columns
bottle_dataset
```

```python
sns.pairplot(bottle_dataset, hue="Bottle_Material")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
rescaled_bottle_dataset = pd.DataFrame(MinMaxScaler().fit_transform(bottle_dataset))
rescaled_bottle_dataset.columns = bottle_dataset_columns
rescaled_bottle_dataset.info()
```

```python
rescaled_bottle_dataset.isna().sum()
```

```python
sns.pairplot(rescaled_bottle_dataset, hue='Bottle_Material')
```

```python
X_train, X_test, y_train, y_test = train_test_split(bottle_dataset.drop(['Bottle_Material'], axis=1), bottle_dataset["Bottle_Material"], test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
baseline_bottle_trainer = DecisionTreeClassifier()
baseline_bottle_model = baseline_bottle_trainer.fit(X_train, y_train)
```

```python
tree.plot_tree(baseline_bottle_model)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import confusion_matrix
```

```python
y_pred = baseline_bottle_model.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python

print("Classification report \n=======================")
print(classification_report(y_true=y_test, y_pred=y_pred))
print("Confusion matrix \n=======================")
print(confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
bottle_1st_trainer = DecisionTreeClassifier(max_depth=4)
bottle_1st_model = bottle_1st_trainer.fit(X_train, y_train)
```

```python
tree.plot_tree(bottle_1st_model)
```

```python
y_1st_pred = bottle_1st_model.predict(X_test)
```

```python
confusion_matrix(y_test, y_1st_pred)
```

```python

print("Classification report \n=======================")
print(classification_report(y_true=y_test, y_pred=y_1st_pred))
print("Confusion matrix \n=======================")
print(confusion_matrix(y_true=y_test, y_pred=y_1st_pred))
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
solvers = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
penalty = ['l1', 'l2', 'elasticnet', None]
c_values = [100, 10, 1.0, 0.1, 0.01]
```

```python
grid = dict(solver=solvers,penalty=penalty, C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
linear_1st_trainer = GridSearchCV(estimator=baseline_linear_trainer, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
```

```python
linear_1st_model = linear_1st_trainer.fit(X_train, y_train)
```

```python
y_linear_1st_pred = linear_1st_model.predict(X_test)
```

```python
confusion_matrix(y_test, y_linear_1st_pred)
```

```python

print("Classification report \n=======================")
print(classification_report(y_true=y_test, y_pred=y_linear_1st_pred))
print("Confusion matrix \n=======================")
print(confusion_matrix(y_true=y_test, y_pred=y_linear_1st_pred))
```
