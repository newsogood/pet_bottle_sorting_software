```python
def bottle_data_optimizer(data, label=None, mat=None):
    count = 0
    for index_0 in range(len(data)):
        for index_1 in range(len(data[index_0])):
            check = list(data[index_0][index_1])
            for letter in data[index_0][index_1]:
                if letter.isalpha():
                    check.remove(letter)
                if ((letter == "[") or (letter == "]")):
                    check.remove(letter)
                if (letter == "."):
                    count += 1
                if ((letter == "⸮") or (letter == "&") or (letter == "_")):
                    count += 1
                    if count == 1:
                        check[check.index(letter)] = "."
                    else:
                        check.remove(letter)
            count = 0
            data[index_0][index_1] = float("".join(check))
        data[index_0].extend([label, mat])
    output = pd.DataFrame(data)
    return output
    
```

```python
output_pet_mat_label = bottle_data_optimizer(pet_mat_label_list, label=1, mat=1)
output_pet_mat = bottle_data_optimizer(pet_mat_list, label=0, mat=1)
output_other_mat_label = bottle_data_optimizer(other_mat_label_list, label=1, mat=0)
output_other_mat = bottle_data_optimizer(other_mat_list, label=0, mat=0)
```

```python
bottle_df = pd.concat([output_other_mat, output_other_mat_label, output_pet_mat, output_pet_mat_label], axis=0)
bottle_df.reset_index(inplace=True)
bottle_df.info()
```
