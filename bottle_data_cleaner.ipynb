```python
print(len(pet_mat_list))
print(len(pet_mat_label_list))
print(len(other_mat_list))
print(len(other_mat_label_list))
```

```python
def bottle_data_optimizer(data, label, mat):
    count = 0
    for index_0 in range(len(data)):
        for index_1 in range(len(data[index_0])):
            check = list(data[index_0][index_1])
            for letter in data[index_0][index_1]:
                if letter.isalpha():
                    check.remove(letter)
                if ((letter == "[") or (letter == "]")):
                    check.remove(letter)
                if (letter == "."):
                    count += 1
                if ((letter == "⸮") or (letter == "&") or (letter == "_")):
                    count += 1
                    if count == 1:
                        check[check.index(letter)] = "."
                    else:
                        check.remove(letter)
            count = 0
            data[index_0][index_1] = float("".join(check))
        data[index_0].extend([label, mat])
    return pd.DataFrame(pet_mat_label_list)
    
```

```python
output_pet_mat_label = bottle_data_optimizer(pet_mat_label_list, 1, 0)
output_pet_mat = bottle_data_optimizer(pet_mat_list, 0, 0)
output_other_mat_label = bottle_data_optimizer(other_mat_label_list, 1, 1)
output_other_mat = bottle_data_optimizer(other_mat_list, 0, 1)
```

```python
bottle_df = pd.concat([output_other_mat, output_other_mat_label, output_pet_mat, output_pet_mat_label], axis=0)
bottle_df.info()
```
